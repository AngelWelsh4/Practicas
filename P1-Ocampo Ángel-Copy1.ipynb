## Diversidad de Rango

El siguiente archivo contiene las 2500 palabras mas usadas de los 50 libros anteriores.  Esste archivo esta estructurado en una lista de listas, donde Palabras[0] corresponde a las palabras mas usadas en el libro 1, Palabras[1] corresponde a las palabras mas usadas en el libro 2,   y asi sucesivamente.

1- Construye un algoritmo para encontrar la diversidad de rango del listado anterior. Te sugerimos primero encontrar el valor de diversidad de cada rango, y al final graficar los rangos vs su diversidad.

Para la grafica usa markers redondos (como un scatter plot) y usa escala logaritmica en el eje x.


```python
Palabras = load("libros/Palabras.npy").tolist()
```

```python
len(Palabras) ##el número de libros en Palabras
```

```python
len(Palabras[0]) ##el número de palabras en el libro 0
```

```python
def dr(l,ll,tr):
    DK=[]
    for i in range(0,tr):
        pal=[]
        for j in range(0,ll):
            palabra= Palabras[j][i]
            if palabra not in pal:
                pal.append(palabra)
        dk=len(pal)/ll
        DK.append(dk)
    return DK
print(dr(Palabras,len(Palabras),len(Palabras[0]))) #Obtenemos el listado de diversidad de rangos
```

```python
r=arange(1,len(Palabras[0])+1,1)
rcParams['figure.figsize']=[7,7]
plot(r, dr(Palabras,len(Palabras),len(Palabras[0])), marker='o', markersize=3)
xscale('log')
xlabel('rango de palabras k')
ylabel('diversidad de rango asociada d(k)')
title('k vs. d(k)')
```

## Extra 1

La curva de diversidad puede ser descrita por la siguiente expresion (dependiente de $\mu$ y $\sigma$).

$$d(k) = \frac{1}{\sigma\sqrt{2\pi}}\int_{-\infty }^{\log_{10^{k}}} e^{-\frac{\left ( y-\mu \right )^{2}}{2\sigma^{2}}}dy.$$ 	


La siguiente funcion devuelve los rangos y sus evaluaciones en la integral anterior.  

A partir de los valores que obtuviste de la diversidad, evalualos en la funcion curva_ms para realizar su ajuste.

Prueba con valores de $\mu = [1, 2]$ y $\sigma = [0.1, 2.1]$. Encuentra la pareja de mu y sigma que mejor ajuste.



```python
from scipy import integrate

def curva_ms(diversidad, mu, sigma):
    K = arange(0,len(diversidad))+1
    val = []
    def fsig(x):
        c = 1 /(sigma*sqrt(2*pi)) 
        y = exp( (-(x-mu)**2) / (2* sigma**2) ) 
        return y*c
    for i in range(0,len(K)):
        k = K[i]
        if log10(k) == 0:
            b = log10(2)/2
        elif log10(k) >0:
            b = log10(k)
        res = integrate.quad(fsig, -inf, b)
        val.append(res[0])
    return K, val
```
